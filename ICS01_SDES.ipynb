```python
P10 = (3, 5, 2, 7, 4, 10, 1, 9, 8, 6)
P8 = (6, 3, 7, 4, 8, 5, 10, 9)
P4 = (2, 4, 3, 1)

IP = (2, 6, 3, 1, 4, 8, 5, 7)
IPi = (4, 1, 3, 5, 7, 2, 8, 6)

E = (4, 1, 2, 3, 2, 3, 4, 1)

S0 = [
        [1, 0, 3, 2],
        [3, 2, 1, 0],
        [0, 2, 1, 3],
        [3, 1, 3, 2]
     ]

S1 = [
        [0, 1, 2, 3],
        [2, 0, 1, 3],
        [3, 0, 1, 0],
        [2, 1, 0, 3]
     ]

def permutation(pattern, key):
    permuted = ""
    for i in pattern:
        permuted += key[i-1]
    return permuted

def generate_k1(left, right):
    left = left[1:] + left[:1]
    right = right[1:] + right[:1]
    key = left+right
    return permutation(P8, key)

def generate_k2(left, right):
    left = left[3:] + left[:3]
    right = right[3:] + right[:3]
    key = left + right
    return permutation(P8, key)
 
def transform(right, key):
    extended = permutation(E, right)
    xor_cipher = bin(int(extended, 2) ^ int(key, 2))[2:].zfill(8)
    xor_left = xor_cipher[:4]
    xor_right = xor_cipher[4:]
    new_left = str(Sbox(xor_left, S0))
    new_right = str(Sbox(xor_right, S1))
    print(new_left+new_right)
    print("Transform : ", permutation(P4, new_left + new_right))
    return permutation(P4, new_left + new_right)

def encrypt(left, right, key):
    cipher = int(left, 2) ^ int(transform(right, key), 2)
    print("Cipher= ", bin(cipher)[2:].zfill(4))
    return right, bin(cipher)[2:].zfill(4)

def dencrypt(left, right, key):
    plaintext=int(left,2)^int(transform(right, key),2)
    return right, bin(plaintext)[2:].zfill(4)

def Sbox(data, box):
    row = int(data[0] + data[3], 2)
    col = int(data[1] + data[2], 2)
    print("Sbox: ", bin(box[row][col])[2:].zfill(4))
    return bin(box[row][col])[2:].zfill(2)  #zfill 2 mistake 1


if __name__=="__main__":
    plaintext = str(10010111)
    # plaintext = input("Enter 8-bit plain text : ")
    if len(plaintext) != 8:
        raise Exception("Invalid plaintext! Please enter 8-bit plaintext")
    key = str(1010000010)
    # key = input("Enter 10-bit key : ")
    if len(key) != 10:
        raise Exception("Invalid key! Please enter 10-bit key")

    p10key = permutation(P10, key)
    print("First Permutation:", p10key)

    left_key = p10key[:len(p10key)//2]
    right_key = p10key[len(p10key)//2:]
    print("Left key : ", left_key)
    print("Right key : ", right_key)

    k1 = generate_k1(left_key, right_key)
    print("\nFirst key:", k1)

    k2 = generate_k2(left_key, right_key)
    print("Second key : ", k2)

    initial_permutation = permutation(IP, plaintext)
    print("\nInitial permutation : ", initial_permutation)

    left_data = initial_permutation[:len(initial_permutation)//2]
    right_data = initial_permutation[len(initial_permutation)//2:]

    left, right = encrypt(left_data, right_data, k1) 
    print(left, right)
    left, right = encrypt(left, right, k2)
    print("Ciphertext:", permutation(IPi, right + left ))
    
    cipherText=permutation(IPi, right + left)
    initial_permutation = permutation(IP, cipherText)
    print("\nInitial permutation : ", initial_permutation)

    left_data = initial_permutation[:len(initial_permutation)//2]
    right_data = initial_permutation[len(initial_permutation)//2:]

    left, right = dencrypt(left_data, right_data, k2) 
    print(left, right)
    left, right = dencrypt(left, right, k1)
    print("Plaintext:", permutation(IPi, right + left ))
```
